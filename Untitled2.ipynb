To Predict the percentage of marks of the students based on the number of hours they studied

```r
#loading given file usind data frame

data <- data.frame(Hours=c(2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.8, 6.9, 7.8),
                   Scores=c(21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 30, 54, 35, 76, 86))
data
```

```r
#Step-2 Plotting the Given Data set.**
    # Plotting the given data
    plot(x = data$Hours, y = data$Scores, xlab = "Hours", ylab = "Scores", main = "Score of student by study Hours", col = "Red")
  
```

```r
#Step-3 Running Linear Regression on the data as there are only Two variables.**
    # Linear regression
   data_regression <- lm(formula = Scores~Hours , data = data)
  plot(x = data$Hours, y = data$Scores, xlab = "Hours", ylab = "Scores", main = "Score of student by study Hours", col = "red")
  abline(data_regression, col= "Blue")
   summary(data_regression)
```

```r
#predicting the score if a student studies for 9.25 hrs/ day?
a<-data.frame(Hours=9.25)
result<-predict(data_regression,a)
print(result)
```
